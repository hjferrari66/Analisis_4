```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
z, t, f = symbols('z t f')
a = symbols('a', real = True)
init_printing(use_unicode=True)
```

```python
z = t + I * a * sin(t)         # parametrizo la curva sobre la que integro
```

```python
diff(z,t)                      # derivada de z(t)
```

```python
f = re(z**2)                   # calculo f(z(t))
```

```python
solucion=integrate(f*diff(z,t), (t, 0, pi))    # calculos las integrales en la variable real t entre a y b
solucion
```

```python
x = np.linspace(0, np.pi, 100)        # pasos para recorrer la curva parametrizada entre a y b, 0 y pi en este caso
```

```python
for k in range(-3, 4):
    plt.plot(x ,k * np.sin(x))
    plt.arrow(x[50], k * np.sin(x[50]), 0.01, k * np.sin(x[51]) - k * np.sin(x[50]), shape='full', lw=0, length_includes_head=True, head_width=.15)
    print('a = ', k, '; integral = ',solucion.evalf(subs={a:k}))
```
